```python
from nltk.corpus import stopwords
stop=stopwords.words('english')
import string
punc=list(string.punctuation)
```

```python
stop=stop+punc
```

```python
#we will create a list which will contain tuples containing the words in a array alog with their category
docs=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        docs.append((movie_reviews.words(fileid),category))
docs        
```

```python
import random
random.shuffle(docs)#shuffling the docs
```

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
```

```python
from nltk import pos_tag
w="better"
pos_tag(w)
#the problem here is pos_tag expects an array as an input
```

```python
w="better"
pos_tag([w])
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):#adverb
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    o=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            new_word=lem.lemmatize(w,pos=get_simple_pos(pos[0][1]))#pos[0][1] as pos_tag gives a 
                                                                   #tuple inside an array
            o.append(new_word.lower())
    return o
```

```python
clean_docs=[]
for i in docs:
    new_r=clean_review(i[0])
    clean_docs.append((new_r,i[1]))
```

```python
print(clean_docs[5])
```

```python
#to use these data as features, mark a word as a positiv eor negative so 2000 docs will make a array of 2000
#tuples where each tuple will have a dictionary containing pos or neg corresponding to every word of that 
#document and the category the most occurring words will be our features
```

```python
train_docs=clean_docs[:1500]
test_docs=clean_docs[1500:]
```

```python
train_docs[0][0][0]
```

```python
all_words=[]
for doc in train_docs:
            all_words+=doc[0]
```

```python
freq=nltk.FreqDist(all_words)#gives frequency of each word like in a dictionary
freq
```

```python
common=freq.most_common(3000)#gives top 3000 words
common
```

```python
features=[i[0] for i in common]
features
```

```python
#now we create a dictinary telling if the feature is present in a particular document or not
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in features:
        current_features[w]=w in words_set
    return current_features
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in train_docs]
```

```python
training_data[0]
```

```python
testing_data=[(get_feature_dict(doc),category) for doc,category in test_docs]
```

```python
from nltk import NaiveBayesClassifier#this classifier requires data to be in the above format
```

```python
classifier=NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)#speaks for itself
```

```python
classifier.show_most_informative_features(15)
#this shows the features which on an average had the most affect on the result, it also gives a ratio of
#neg:pos or pos:neg as to show the ratio of probability of a document to be a neg or a pos when that 
#particular feature was true
```

## n_grams- the score can be improved by using another parameter ot CountVectorizer i.e. n_gram, n here is how many words can be used in a single feature, default is 1, for ex:-"today is  not good" has 3 bigrams "today is","is not","not good", good as a feature gives a different emotion compared to not good which is completely opposite

```python
cvc=CountVectorizer(max_features=2000,ngram_range=(1,3))#this means i want to take 
                                                        #1 grams bi-grams and tri-grams
x_train_features=cvc.fit_transform(x_train)
x_train_features.todense()
```

## TF-IDF---tf and idf mean 'term frequency' and 'inverse document frequency' respectively, document frequency of a word means how in many documents the word appears, for ex - if a word appears once in more than 80% of the docs then though its frequency is high it will not be much use to us as a feature as comparison is the key rather a word appering in 10% of the documents but with the same total frequency as the the previous word is more useful when comparing

## CountVectorizer has arguments for this min_df and max_df which takes values between 0 and 1, if max_df=0.7 then we dont take a word that is present in more than 70% of the docs

## tf is just any cell in our feature array like how many times wi ocuurs in di

## min_df and max_df are only useful when selcting our top features but in our actual array instead of just using tf we can use tf*idf which lowers the weightage of a word appearing in multiple docs

## idf is actually 1/df, and there is a tfidfvectorizer for this kind of build as well
